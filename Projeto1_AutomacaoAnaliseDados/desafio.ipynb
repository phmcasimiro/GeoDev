### AUTOMAÇÃO DE ACESSO E DOWNLOAD DE DADOS

```python
#--------------------------------#
# ETAPA 1 - ENTRANDO NO SISTEMA
#--------------------------------#

import pyautogui
import time

# Tempo padrão entre todos os comandos do pyautogui
pyautogui.PAUSE = 1

# Pressionar a tecla do Windows
pyautogui.press('win')

# Digitar Edge
pyautogui.write('edge')

# Apertar enter para abrir o Edge
pyautogui.press('enter')
time.sleep(3) # Aguardar 3 segundos após o enter

# definição de endereço em uma variável link
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"

# Digitar o site e apertar enter
pyautogui.write(link)
pyautogui.press('enter')
time.sleep(5) # Aguardar 5 segundos após o enter
```

```python
#-------------------------------#
# ETAPA 2 - NAVEGANDO NO SISTEMA
#-------------------------------#

# Duplo click na pasta do drive
pyautogui.click(x=356, y=385, clicks=2)
time.sleep(2) # Aguardar 2 segundos após o enter

#-------------------------------#
# ETAPA 3 - DOWNLOAD DA PLANILHA
#-------------------------------#

# OPÇÃO 1 - Clicar em 3 pontos para abrir a lista de ações
pyautogui.click(x=925, y=382)
time.sleep(5) # Aguardar 2 segundos após o enter

# OPÇÃO 1 - Clicar em transferir/download
pyautogui.click(x=967, y=347)
time.sleep(15) # Aguardar 2 segundos após o enter

'''
A OPÇÃO 2 tem o seguinte problema, se você não clicar exatamente no mesmo lugar, haverá deslocamento da caixa de ações

# OPÇÃO 2 - Clicar com botão direito no arquivo
pyautogui.click(x=925, y=382, button='right')
time.sleep(5) # Aguardar 2 segundos após o enter

# OPÇÃO 2 - Clicar em transferir/download
pyautogui.click(x=505, y=346)
time.sleep(15) # Aguardar 2 segundos após o enter
'''
```
