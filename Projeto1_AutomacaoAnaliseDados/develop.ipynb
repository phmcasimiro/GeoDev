### INGESTÃO, TRATAMENTO E ANÁLISE DE DADOS
### Analisar Dados

- ETAPA 1: ABRIR A BASE DE DADOS
- ETAPA 2: AVALIAR AS COLUNAS DE INFORMAÇÕES DA BASE DE DADOS
- ETAPA 3: CONSTRUIR UMA ANÁLISE DESCRITIVA DOS PRODUTOS
    - Faturamento = Soma da coluna "Valor Final"
    - Quantidade de Produtos = Soma da coluna "Quantidade"


```python
import pandas as pd
import os

caminho = r"C:\Users\HP\Downloads\Vendas - Dez.xlsx"

# 1. Importa os dados
ds_1 = pd.read_excel(caminho)

# 2. Deleta o arquivo após a importação
if os.path.exists(caminho):
    os.remove(caminho)
    print("Arquivo deletado com sucesso!")
else:
    print("O arquivo não foi encontrado.")

ds_1
```

```python
faturamento = ds_1['Valor Final'].sum()

qtd_produtos = ds_1['Quantidade'].sum()

print(faturamento)
print(qtd_produtos)
```

```python
display(ds_1.describe())
```

```python
valor_total_vendido = ds_1['Valor Final'].sum()


#--------------------------------------------------------------------#
# Vendas com maior quantidade de itens
#--------------------------------------------------------------------#

# Agrupar por 'Código Venda' e somar a 'Quantidade' de cada grupo
vendas_por_codigo = ds_1.groupby('Código Venda')['Quantidade'].sum()
# Encontrar o valor máximo dessa soma
maior_quantidade = vendas_por_codigo.max()

# Filtrar a série para pegar apenas os itens onde a quantidade é igual à maior_quantidade
vendas_maximas = vendas_por_codigo[vendas_por_codigo == maior_quantidade].index
# Transformar esses códigos em uma lista e depois em uma string separada por vírgulas
lista_codigos_vendas_maximas = ','.join(map(str,vendas_maximas))

#--------------------------------------------------------------------#
# Vendas com menor quantidade de itens
#--------------------------------------------------------------------#

# Encontrar o valor mínimo dessa soma
menor_quantidade = vendas_por_codigo.min()
# Filtrar a série para pegar apenas os itens onde a quantidade é igual à menor_quantidade
vendas_minimas = vendas_por_codigo[vendas_por_codigo == menor_quantidade].index
# Transformar esses códigos em uma lista e depois em uma string separada por vírgulas
lista_codigos_vendas_minimas = ", ".join(map(str, vendas_minimas))

#--------------------------------------------------------------------#
# Vendas com maior valor
#--------------------------------------------------------------------#

# Agrupar por 'Código Venda' e somar o 'valor' de cada venda
vendas_por_valor = ds_1.groupby('Código Venda')['Valor Final'].sum()
# Encontrar o valor máximo dessa soma
maior_valor = vendas_por_valor.max()
# Filtrar a série para pegar apenas as vendas com maior valor
codigos_vendas_mais_caras = vendas_por_valor[vendas_por_valor == maior_valor].index
# Transformar esses códigos em uma lista e depois em uma string separada por vírgulas
lista_codigos_vendas_mais_caras = ','.join(map(str,codigos_vendas_mais_caras))


#--------------------------------------------------------------------#
# Vendas com menor valor
#--------------------------------------------------------------------#

# Encontrar o valor máximo dessa soma
menor_valor = vendas_por_valor.min()
# Filtrar a série para pegar apenas as vendas com menor valor
codigos_vendas_mais_baratas = vendas_por_valor[vendas_por_valor == menor_valor].index
# Transformar esses códigos em uma lista e depois em uma string separada por vírgulas
lista_codigos_vendas_mais_baratas = ','.join(map(str,codigos_vendas_mais_baratas))

#--------------------------------------------------------------------#
# Média dos valores das vendas
#--------------------------------------------------------------------#

# Agrupar por código e somar o valor final de cada venda
valor_por_venda = ds_1.groupby('Código Venda')['Valor Final'].sum()
# Calcular a média dessas vendas totais
media_vendas = valor_por_venda.mean().round(2)

#--------------------------------------------------------------------#
# Moda dos valores das vendas
#--------------------------------------------------------------------#

# Calcular a moda das vendas totais
moda_vendas = valor_por_venda.mode().round(2)

#--------------------------------------------------------------------#
# Mediana dos valores das vendas
#--------------------------------------------------------------------#

# Calcular a moda das vendas totais
mediana_vendas = valor_por_venda.median().round(2)

#--------------------------------------------------------------------#
# Desvio Padrão dos valores das vendas
#--------------------------------------------------------------------#

# Calcula o desvio padrão:
desvio_padrao_vendas = valor_por_venda.std()


print(f'O valor total das vendas foi R$ {valor_total_vendido:,.2f}')
print(f'A quantidade de vendas foi: {ds_1['Código Venda'].nunique()}') # Quantidade de vendas
print(f"A maior quantidade de produtos em uma venda foi: {maior_quantidade}")
print(f"O código da venda correspondente é: {lista_codigos_vendas_maximas}")
print(f'A venda de maior valor foi R$ {lista_codigos_vendas_mais_caras}')
print(f"A menor quantidade de produtos foi: {menor_quantidade}")
print(f"Os códigos das vendas correspondentes são: {lista_codigos_vendas_minimas}")
print(f'A venda de menor valor foi: {lista_codigos_vendas_mais_baratas}')
print(f'O maior valor em vendas foi de R$ {maior_valor:,.2f}')
print(f'O menor valor em vendas foi de R$ {menor_valor:,.2f}')
print(f'A média dos valores das vendas (Ticket médio) foi de R$ {media_vendas:,.2f}')
print(f'A moda dos valores das vendas foi de R$ {moda_vendas[0]:,.2f}')
print(f'A mediana dos valores das vendas foi de R$ {mediana_vendas:,.2f}')
print(f"O desvio padrão das vendas foi de: R$ {desvio_padrao_vendas:,.2f}")
```

### Leitura do Gráfico

1. A Caixa Central

- A caixa representa onde estão 50% dos seus dados. É a maior parte das vendas.
- A base da caixa (Q1 - 1º Quartil - R$ 262.00): 25% das vendas estão abaixo desse valor.
- O topo da caixa (Q3 - 3º Quartil - R$ 1,042.50): 75% das vendas estão abaixo desse valor.
- Análise: A caixa é pequena, ou seja, os valores das vendas são muito parecidos.
- A Mediana (linha do meio da caixa) divide seus dados exatamente ao meio.
- 50% das vendas foram maiores e 50% foram menores que a mediana.
- Diferença da Média: Como a linha da mediana não está exatamente no meio da caixa, mas está mais para baixo, significa que os dados são "tendenciosos" (assimétricos). Isso significa que há mais vendas de menor valor e que a média é elevada pelas vendas de maior valor e que são mais raras. Isso é corroborado no gráfico de frequência (Quantidade de Vendas e Valor das Vendas)
- As Hastes (Whiskers) indicam a dispersão geral dos dados/vendas, isto é, representam o limite do que é considerado um valor "normal" no conjunto de dados. Os dados/vendas que estiverem dentro dessas linhas são considerados comportamento padrão das vendas.
- Qualquer ponto que aparecer fora das hastes é um Outlier (um valor fora da curva). Na prática são vendas excepcionais, isto é, um cliente que comprou muito mais que o normal ou um erro de digitação.
