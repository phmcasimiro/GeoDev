## CHEET SHEAT DO PROJETO

##### Guia Rápido de Referência: PyAutoGUI

| Categoria | Comando | O que faz? | Dica de Erro Comum (Bug Alert ⚠️) |
| :--- | :--- | :--- | :--- |
| **Configuração** | `pyautogui.PAUSE` | Define pausa entre comandos. | **Esquecer de configurar:** Sem isso, o robô tenta "atropelar" o Windows e nada funciona. |
| **Mouse** | `pyautogui.click()` | Clica na tela. | **Mudança de resolução:** Se você mudar de monitor, as coordenadas (X, Y) mudam e o robô clica no lugar errado. |
| **Mouse** | `pyautogui.doubleClick()` | Clique duplo. | **Velocidade:** Às vezes o Windows não entende o duplo clique se o PC estiver lento; adicione um `time.sleep` antes. |
| **Mouse** | `pyautogui.rightClick()` | Botão direito. | **Menu contextual:** O menu pode demorar a abrir. Não dê o próximo comando imediatamente. |
| **Mouse** | `pyautogui.moveTo()` | Move o cursor. | **Caminho obstruído:** Tentar mover o mouse enquanto o usuário também está mexendo gera conflito. |
| **Teclado** | `pyautogui.write()` | Digita texto. | **Teclado ABNT2/Estrangeiro:** Caracteres especiais (ç, ~, á) podem sair errados se o layout do teclado for diferente. |
| **Teclado** | `pyautogui.press()` | Aperta uma tecla. | **Nome da tecla:** Escrever `press("ENTER")` em maiúsculo pode dar erro em algumas versões. Use sempre minúsculo: `"enter"`. |
| **Teclado** | `pyautogui.hotkey()` | Atalhos (Ctrl+C). | **Ordem das teclas:** Se colocar `("c", "ctrl")` em vez de `("ctrl", "c")`, o atalho pode não funcionar. |
| **Teclado** | `pyautogui.keyDown()` | Segura uma tecla. | **Esquecer de soltar:** Se você usar `keyDown` e não usar o `keyUp` depois, seu teclado vai ficar "travado" naquela tecla. |

---

##### **Dicas**

- **Segurança (Fail-Safe):** Se o seu robô ficar "maluco", leve o mouse rapidamente para o **canto superior esquerdo da tela**. Isso forçará o programa a parar imediatamente.
- **Atenção ao Tempo:** Sempre use `time.sleep()` ou o `pyautogui.PAUSE`. O computador é muito mais rápido que o carregamento dos sites; sem pausas, o robô tentará clicar em botões que ainda não apareceram.
- **Coordenadas:** Lembre-se que a coordenada `(0,0)` é o topo esquerdo da sua tela.
- Se o comando `pyautogui.write()` não estiver funcionando bem com acentos ou cedilha, a melhor prática profissional é:
    - Copiar o texto para a área de transferência (usando a biblioteca `pyperclip`).
    - Usar o `pyautogui.hotkey('ctrl', 'v')` para colar.

```python
#-------------------------------#
# ETAPA 4 - INGESTÃO
#-------------------------------#

import pandas as pd
import os

caminho = r"C:\Users\HP\Downloads\Vendas - Dez.xlsx"

# Importar dados
ds_1 = pd.read_excel(caminho)

# Deletar o arquivo após a importação
if os.path.exists(caminho):
    os.remove(caminho)
    print("Arquivo deletado com sucesso!")
else:
    print("O arquivo não foi encontrado.")

# Exibir o dataframe
ds_1
```

```python
#-------------------------------#
# ETAPA 5 - AVALIAÇÃO PRELIMINAR DO DATAFRAME
#-------------------------------#
print('#----------------------------------------------------#')
print('LINHAS & COLUNAS')
print('#----------------------------------------------------#')
print(ds_1.shape)

print('#----------------------------------------------------------#')
print('AVALIAÇÃO PRELIMINAR - ESTATÍSTICA DESCRITIVA AUTOMATIZADA')
print('#----------------------------------------------------------#')
display(ds_1.describe())
```

```python
#-------------------------------#
# ETAPA 6 - ANÁLISE DE DADOS (ESTATÍSTICA DESCRITIVA)
#-------------------------------#

#--------------------------------------------------------------------#
# VALOR TOTAL DE VENDAS
#--------------------------------------------------------------------#
valor_total_vendido = ds_1['Valor Final'].sum()

#--------------------------------------------------------------------#
# AGRUPAMENTO DE 'Produto' POR  'Código Venda' PARA SABER A 'Quantidade' DE ITENS POR VENDA 
#--------------------------------------------------------------------#
vendas_por_codigo = ds_1.groupby('Código Venda')['Quantidade'].sum()

#--------------------------------------------------------------------#
# VENDA COM MAIOR QUANTIDADE DE 'Produto"
#--------------------------------------------------------------------#
maior_quantidade = vendas_por_codigo.max()

#--------------------------------------------------------------------#
# VENDAS ('Código Venda') COM MAIOR QUANTIDADE DE ITENS ('Produto')
# SELEÇÃO DAS VENDAS ('vendas_por_codigo') COM MAIOR QUANTIDADE ('maior_quantidade')
#--------------------------------------------------------------------#
vendas_maximas = vendas_por_codigo[vendas_por_codigo == maior_quantidade].index

#--------------------------------------------------------------------#
# LISTAR CÓDIGOS DAS MAIORES VENDAS EM UMA STRING SEPARADA POR VÍRGULAS
#--------------------------------------------------------------------#
lista_codigos_vendas_maximas = ','.join(map(str,vendas_maximas))

#--------------------------------------------------------------------#
# VENDA COM MENOR QUANTIDADE DE 'Produto"
#--------------------------------------------------------------------#
menor_quantidade = vendas_por_codigo.min()

#--------------------------------------------------------------------#
# VENDAS ('Código Venda') COM MENOR QUANTIDADE DE ITENS ('Produto')
# SELEÇÃO DAS VENDAS ('vendas_por_codigo') COM MENOR QUANTIDADE ('maior_quantidade')
#--------------------------------------------------------------------#
vendas_minimas = vendas_por_codigo[vendas_por_codigo == menor_quantidade].index

#--------------------------------------------------------------------#
# LISTAR CÓDIGOS DAS MENORES VENDAS EM UMA STRING SEPARADA POR VÍRGULAS
#--------------------------------------------------------------------#
lista_codigos_vendas_minimas = ", ".join(map(str, vendas_minimas))

#--------------------------------------------------------------------#
# VENDAS DE MAIOR VALOR
#--------------------------------------------------------------------#
# AGRUPAR POR 'Código Venda' E SOMAR O 'valor' DE CADA VENDA
valor_por_venda = ds_1.groupby('Código Venda')['Valor Final'].sum()
# ENCONTRAR O VALOR MÁXIMO DESSA SOMA
maior_valor = valor_por_venda.max()
# SELEÇÃO DAS VENDAS COM MAIOR VALOR
codigos_vendas_mais_caras = valor_por_venda[valor_por_venda == maior_valor].index
# LISTAR CÓDIGOS DAS VENDAS DE MAIOR VALOR EM UMA STRING SEPARADA POR VÍRGULAS
lista_codigos_vendas_mais_caras = ','.join(map(str,codigos_vendas_mais_caras))

#--------------------------------------------------------------------#
# VENDAS DE MENOR VALOR
#--------------------------------------------------------------------#
# ENCONTRAR O VALOR MÍNIMO DESSA SOMA
menor_valor = valor_por_venda.min()
# SELEÇÃO DAS VENDAS COM MENOR VALOR
codigos_vendas_mais_baratas = valor_por_venda[valor_por_venda == menor_valor].index
# LISTAR CÓDIGOS DAS VENDAS DE MENOR VALOR EM UMA STRING SEPARADA POR VÍRGULAS
lista_codigos_vendas_mais_baratas = ','.join(map(str,codigos_vendas_mais_baratas))

#--------------------------------------------------------------------#
# MÉDIA DOS VALORES DAS VENDAS
#--------------------------------------------------------------------#
media_vendas = valor_por_venda.mean().round(2)

#--------------------------------------------------------------------#
# MODA DOS VALORES DAS VENDAS
#--------------------------------------------------------------------#
moda_vendas = valor_por_venda.mode().round(2)

#--------------------------------------------------------------------#
# MEDIANA DOS VALORES DAS VENDAS
#--------------------------------------------------------------------#
mediana_vendas = valor_por_venda.median().round(2)

#--------------------------------------------------------------------#
# DESVIO PADRÃO DOS VALORES DAS VENDAS
#--------------------------------------------------------------------#
desvio_padrao_vendas = valor_por_venda.std()


print(f'O valor total das vendas foi R$ {valor_total_vendido:,.2f}')
print(f'A quantidade de vendas foi: {ds_1['Código Venda'].nunique()}') # Quantidade de vendas
print(f"A maior quantidade de produtos em uma venda foi: {maior_quantidade}")
print(f"O código da venda correspondente é: {lista_codigos_vendas_maximas}")
print(f'A venda de maior valor foi R$ {lista_codigos_vendas_mais_caras}')
print(f"A menor quantidade de produtos foi: {menor_quantidade}")
print(f"Os códigos das vendas correspondentes são: {lista_codigos_vendas_minimas}")
print(f'A venda de menor valor foi: {lista_codigos_vendas_mais_baratas}')
print(f'O maior valor em vendas foi de R$ {maior_valor:,.2f}')
print(f'O menor valor em vendas foi de R$ {menor_valor:,.2f}')
print(f'A média dos valores das vendas (Ticket médio) foi de R$ {media_vendas:,.2f}')
print(f'A moda dos valores das vendas foi de R$ {moda_vendas[0]:,.2f}')
print(f'A mediana dos valores das vendas foi de R$ {mediana_vendas:,.2f}')
print(f"O desvio padrão das vendas foi de: R$ {desvio_padrao_vendas:,.2f}")
```

```python
#----------------------------------------------#
# ETAPA 7 - ELABORAÇÃO DE GRÁFICOS & DIAGRAMAS
#----------------------------------------------#

import plotly.express as px

# Criar um Histograma com Boxplot na margem
fig = px.histogram(
    ds_1, 
    x="Valor Final", 
    nbins=50, # Quantidade de barras (ajustar se necessário)
    title="Análise de Assimetria e Distribuição de Vendas",
    marginal="box", # Adiciona o Boxplot logo acima do histograma!
    color_discrete_sequence=['#636EFA'] # Cor azul profissional
)

# Estilização do gráfico
fig.update_layout(
    xaxis_title="Valor da Venda (R$)",
    yaxis_title="Frequência (Quantidade de Vendas)",
    bargap=0.1 # Espaço entre as barras
)

fig.show()
```
